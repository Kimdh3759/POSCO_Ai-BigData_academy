### 정제(cleaning) 와 정규화 (normalization)

```python
from konlpy.tag import Hannanum
from collections import Counter

tagger = Hannanum()

text= '대한민국은 민주공화국이다. 대한민국의 주권은 국민에게 있고, 국민은 그 대표자나 국민투표에 의하여 주권을 행사한다.'
result = tagger.morphs(text)

c=Counter(result)
print(c.most_common(5))
```

### 불용어 제거(stopword)

```python
from konlpy.tag import Hannanum
from collections import Counter

tagger = Hannanum()

text ='대한민국은 민주공화국이다. 대한민국의 주권은 국민에게 있고, 국민은 그 대표자나 국민투표에 의하여 주권을 행사한다.'
result = tagger.morphs(text)

stop_words = ['은', '.','이','다','의','에게','있','고','그','에','어','을','하','ㄴ다',',']
result = [x for x in result if x not in stop_words]

c= Counter(result)
print(c.most_common(5))
```

### 어근 동일화(stemming) 

```python
from konlpy.tag import Okt
tagger = Okt()
text = '한글 자연어 처리는 재밌다. 이제부터 열심히 해야지'

print(tagger.morphs(text))
print(tagger.morphs(text, stem = True))
```

```python
text = '한다 하겠다 할것이다'
print(tagger.morphs(text,stem=True))
```

```python
from konlpy. tag import Okt
import re

with open('/home/piai/Desktop/POSCO_Ai-BigData_academy/Week4_Webcrolling/실습파일/news.txt','r',encoding = 'utf-8') as f:
    content = f.read()
    
p = re.compile('[\Wa-zA-Z0-9_]+')
content = re.sub(p ,' ', content)

tagger= Okt()
morphs = tagger.pos(content)
print(morphs)
```

```python
words = []
for word, pos in morphs:
    if pos == 'Noun':
        words.append(word)
print(words)
```

#  유사도를 이용한 영화 추천 시스템

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel 

data =pd.read_csv('/home/piai/Desktop/POSCO_Ai-BigData_academy/Week4_Webcrolling/실습파일/movies_metadata.csv',low_memory=False)
data = data.iloc[:20000]
data.head()
```

```python
data['overview']=data['overview'].fillna('')
print(data.shape)
```

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(data['overview'])
print(tfidf_matrix.shape)
```

```python
cosine_sim = linear_kernel(tfidf_matrix,tfidf_matrix)
print(cosine_sim)
```

```python
cosine_sim = linear_kernel(tfidf_matrix,tfidf_matrix)
print(cosine_sim[0][:10])

title = "The Dark Knight Rises"
title2index = pd.Series(data=data.index, index=data['title']).drop_duplicates()
index= title2index[title]

sim_scores = list(enumerate(cosine_sim[index]))

sim_scores.sort(key=lambda x:x[1], reverse= True)
top10 = sim_scores[1:11]
top10_index = [x[0] for x in top10]
print(top10_index)
print(data['title'].iloc[top10_index])
```
